```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2, 2, 100)
x
```

```python
phi1 = np.exp(-2*(x+1)**2)
phi2 = np.exp(-2*(x)**2)
phi3 = np.exp(-2*(x-1)**2)
```

```python
w = np.random.randn(3)
w
```

```python
plt.plot(x, w[0]*phi1 + w[1]*phi2 + w[2]*phi3, "-")
```

```python
anotherW = np.random.randn(3)
print(anotherW)
plt.plot(x, anotherW[0]*phi1 + anotherW[1]*phi2 + anotherW[2]*phi3, "-")
```
